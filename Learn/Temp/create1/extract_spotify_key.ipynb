```python
import pandas as pd
import json 
import csv
import re
```

```python
with open('challenge_set.txt') as text:
    lines = text.readlines()

print(len(lines))
```

```python
str = ""

for i in range(len(lines)):
    str += lines[i]

print(len(str))
```

```python
pattern = re.findall(r'spotify:track:+[\w\.-]+',str)
print(len(pattern))
```

```python
print(pattern[0])
print(pattern[12993])
```

```python
df = pd.DataFrame(pattern)
df.to_csv('spotify_id.csv')

df[0][0]

df1 = df

for i in range(len(df1)):
    df1[0][i] = df[0][i].replace("spotify:track:","")

df1
df1.to_csv("spotify_id_replaced.csv")
```
